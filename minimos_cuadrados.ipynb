```python
from numpy import array
from numpy import loadtxt
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
from pylab import plot, show

dataco2= loadtxt("co2_mm_gl.txt",float)
X=array(dataco2[:,2])
Y=array(dataco2[:,3])
X1=array(dataco2[0:235,2])
Y1=array(dataco2[0:235,3])
X2=array(dataco2[235:470,2])
Y2=array(dataco2[235:470,3])
plot(X,Y,alpha=1,c="green")
#plot(X1,Y1,alpha=1)
X=array([np.ones(470),X]).T
X1=array([np.ones(235),X1]).T
X2=array([np.ones(235),X2]).T
B=np.linalg.inv(X.T @ X) @ X.T @ Y
B1=np.linalg.inv(X1.T @ X1) @ X1.T @ Y1
B2=np.linalg.inv(X2.T @ X2) @ X2.T @ Y2
plot([1980,2019],[B[0]+B[1]*1980,B[0]+B[1]*2019],c="red")
plot([1980,2000],[B1[0]+B1[1]*1980,B1[0]+B1[1]*2000],c="blue")
plot([2000,2019],[B2[0]+B2[1]*2000,B2[0]+B2[1]*2019],c="black")
show()
```

El comportamiento del CO2 en la atmosfera ha crecido de manera significativa en los ultimos 40 años como se puede se puede observar en la grafica verde, el comportamiento oscilatoria que podemos observar entre los intervalos de años es muy parecido al comportamiendo de un |sen(x)|, el comportamiento omitiendo el comportamiento oscilatorio parece lienal, sin embargo , podemos notar que la linealizacion con la recta de color rojo en los extremos se va alejando por lo que podemos predecir que el comportamiento en los proximos años no sera lineal, sino que tendera mas a ser una funcion creciente no lineal, con el analisis que acabamos de hacer, nos permitira dar un modelo bastante burdo podriamos decir que es el siguiente:
\begin{equation}
y=|a*sin(x*b)|+ c* x^{n}
\end{equation}
a conticuacion se muestra un ejemplo del tipo de funcion que podria representar de una manera aproximada el comportamiento de dicho fenomeno.

```python
from numpy import linspace, sin
import math
from pylab import plot, show
a=200
b=3
n=2
c=1/40
x= linspace(1980, 2019,400)
y= abs(a*sin(b*x))+(pow(x,n))*c
plot (x,y)
show()
```
